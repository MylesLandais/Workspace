# Party Character Sheet Viewer

This notebook demonstrates how to view and manage party character sheets using Neo4j queries.


```python
import sys
from pathlib import Path

# Add project root to path
project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

from rpg_graph_vtt.graph.connection import get_connection
from rpg_graph_vtt.graph.queries import PartyQueries, CharacterQueries

# Connect to Neo4j
conn = get_connection()
print("Connected to Neo4j")
```

## Create a Party

First, let's create a party and add some characters to it.


```python
# Create a party
party_name = "The Fellowship"
party_data = {
    "name": party_name,
    "description": "A group of adventurers on an epic quest"
}

try:
    created_party = PartyQueries.create_party(party_data, conn)
    print(f"✓ Created party: {created_party}")
except Exception as e:
    print(f"Party may already exist: {e}")
```

## View All Parties

List all parties in the database.


```python
# Get all parties
parties = PartyQueries.get_all_parties(conn)

print("All Parties:")
for party in parties:
    print(f"\n  {party['name']}")
    print(f"    Characters: {party['character_count']}")
    if party['character_names']:
        print(f"    Members: {', '.join(party['character_names'])}")
```

## View Party Characters

Get detailed information about all characters in a party.


```python
# Get all characters in the party
characters = PartyQueries.get_party_characters(party_name, conn)

print(f"Characters in '{party_name}':\n")
for char in characters:
    print(f"  {char['name']} (Level {char['level']})")
    print(f"    HP: {char['hit_points']}/{char['hit_points_max']} | AC: {char['armor_class']}")
    
    if char.get('classes'):
        classes_str = ', '.join([c['name'] for c in char['classes']])
        print(f"    Class: {classes_str}")
    
    if char.get('races'):
        races_str = ', '.join([r['name'] for r in char['races']])
        print(f"    Race: {races_str}")
    
    if char.get('items'):
        items_str = ', '.join([i['name'] for i in char['items'][:3]])  # Show first 3 items
        if len(char['items']) > 3:
            items_str += f" (+{len(char['items']) - 3} more)"
        print(f"    Items: {items_str}")
    
    print()
```

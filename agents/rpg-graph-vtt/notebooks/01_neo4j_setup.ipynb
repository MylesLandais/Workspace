# Neo4j Setup

Quick setup: Connect to Neo4j, initialize schema, and seed D&D 5e data.


```python
import sys
from pathlib import Path
project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

from rpg_graph_vtt.graph.connection import get_connection
from rpg_graph_vtt.graph.migrations import run_all_migrations, run_migration

# Connect and setup
conn = get_connection()
print(f"✓ Connected to Neo4j")

# Run migrations
migrations_dir = project_root / "neo4j" / "migrations"
run_all_migrations(migrations_dir, conn)

# Seed data
seeds_dir = project_root / "neo4j" / "seeds"
for seed_file in sorted(seeds_dir.glob("*.cypher")):
    run_migration(seed_file, conn)

print("\n✓ Setup complete!")
```

```python
import json
import time
import requests

COMFY_URL = "http://localhost:8188"  # or container host/port

# 1) Load the base workflow JSON exported from Comfy
with open("workflow_api.json", "r") as f:
    workflow = json.load(f)

# 2) Optionally modify prompt / parameters before sending
# Example: if your prompt node is ID "3" and input key is "text"
workflow["3"]["inputs"]["text"] = "a cozy mid-century reading nook, 4k, photorealistic"

# 3) Submit the prompt
resp = requests.post(f"{COMFY_URL}/prompt", json={"prompt": workflow})
resp.raise_for_status()
prompt_result = resp.json()
print(prompt_result)
# Many builds return a prompt_id you can poll for status
prompt_id = prompt_result.get("prompt_id")
```
